```python
f = open('/content/MultiModalClassifier/TorchClassifier/output.txt','r')
acc = []
loss = []
ac = 0
for l in f:
  if l == 'model_loss\n':
    ac = 1
    continue
  else:
    if l == '\n':
      continue
    else:
      if l == 'model_acc\n' or l == '\n':
        continue
      else:
        acc.append(float(l))
  if ac == 1:
    if l == '\n':
      continue
    else:
      loss.append(float(l))
f.close()
print(len(acc))
print(len(loss))
```

```python
import matplotlib.pyplot as plt
plt.plot(acc[0:80:2])
plt.plot(acc[1:80:2])
plt.legend(['training','validation'])
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.title('Accuracy')
plt.show()
```

```python
plt.plot(loss[0:80:2])
plt.plot(loss[1:80:2])
plt.legend(['training','validation'])
plt.xlabel('epoch')
plt.ylabel('loss')
plt.title('Loss')
plt.show()
```
